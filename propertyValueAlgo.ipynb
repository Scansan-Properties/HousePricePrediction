## Score the area based on the schools around it 

```python
postcodes = ['EC1A','EC1M','EC1N','EC1P','EC1R','EC1V','EC1Y','EC2A','EC2M','EC2N','EC2P','EC2R','EC2V','EC2Y','EC2A'
             ,'EC2M','EC2N','EC2P','EC2R','EC2V','EC2Y','EC4A','EC4M']

scoresEdu =[]
for i in postcodes:
    schoolsResponse_API = requests.get('https://api.propertydata.co.uk/schools?key=DCR205CWVJ&postcode='+i)
    
    data = schoolsResponse_API.json()
    raw_data = data.get("data")

    scores = []
    if raw_data!= None:
        raw_data = raw_data.get("state").get("nearest")
        for i in raw_data:
            rating = i.get("rating")
            scores.append(rating)

        scores
        #                                  Calculate education score 
        outstandingCount = scores.count("Outstanding")
        goodCount = scores.count("Good")
        requiresImprovementCount = scores.count("Requires improvement")
        inadequateCount = scores.count("Inadequate")
        unknownCount = scores.count("Unknown") 

        educationScore = (outstandingCount*5) +(goodCount*4)+(requiresImprovementCount*3) +(inadequateCount*0) +(unknownCount*1)
        # educationScore = 20
        classificationScore =" "
        if educationScore >= 20:
            classificationScore = "Great"
        elif 10 <= educationScore < 20:
            classificationScore = "Good"
        else:
            classificationScore = "Poor"

        scoresEdu.append(classificationScore)
    time.sleep(5)
    
    


```
